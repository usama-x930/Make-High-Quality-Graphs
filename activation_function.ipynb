```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
# Set the figure size
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [5, 3]
# Set up with a higher resolution screen
%config InlineBackend.figure_format = 'retina'
plt.style.use('fivethirtyeight')
def relu(z):
    return np.maximum(0, z)

def derivative_relu(z):
    return (z >= 0).astype(z.dtype)

z = np.linspace(-5, 5, 200)

plt.figure()
plt.plot(z, derivative_relu(z), "g-.", linewidth=2, label="ReLU Derivation")
plt.grid(True)
plt.legend(loc='upper left', fontsize=12)
plt.title("Derivative of ReLU Activation Function", fontsize=14)

plt.figure()
plt.plot(z, relu(z), "b" ,linewidth=3, label="ReLU")
plt.grid(True)
plt.legend(loc='upper left', fontsize=12)
plt.title("ReLU Activation Function", fontsize=14)
plt.savefig('relu.png', dpi=500, bbox_inches='tight')
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
# Set the figure size
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [5, 3]
# Set up with a higher resolution screen
%config InlineBackend.figure_format = 'retina'
plt.style.use('fivethirtyeight')
def sig(z):
    return (1 / (1 + np.exp(-z)))

def derivative_sig(z):
    return (z >= 0).astype(z.dtype)

z = np.linspace(-5, 5, 200)

plt.figure()
plt.plot(z, derivative_sig(z), "g-.", linewidth=2, label="Sigmoid Derivation")
plt.grid(True)
plt.legend(loc='upper left', fontsize=12)
plt.title("Derivative of Sigmoid Activation Function", fontsize=14)

plt.figure()
plt.plot(z, sig(z), "b" ,linewidth=3, label="Sigmoid")
plt.grid(True)
plt.legend(loc='upper left', fontsize=12)
plt.title("Sigmoid Activation Function", fontsize=14)
plt.savefig('sigmoid.png', dpi=500, bbox_inches='tight')
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

def sigmoid(z):
    return 1/ (1+np.exp(-z))

def sigmoid_derivation(z):
    return sigmoid(z)*(1-sigmoid(z))

z = np.linspace(-5, 5, 200)
plt.figure()
plt.style.use('fivethirtyeight')
plt.plot(z, sigmoid(z), "b", linewidth=2)
props = dict(facecolor='black', shrink=0.1)
plt.annotate('Saturating', xytext=(3.5, 0.7), xy=(5, 1), arrowprops=props, fontsize=14, ha="center")
plt.annotate('Saturating', xytext=(-3.5, 0.3), xy=(-5, 0), arrowprops=props, fontsize=14, ha="center")
plt.title("Sigmoid Activation Function", fontsize=14)
plt.savefig('sigmoid.png')

z = np.linspace(-5, 5, 200)
plt.figure()
plt.style.use('fivethirtyeight')
plt.plot(z, sigmoid_derivation(z), "b-.", linewidth=2)
props = dict(facecolor='black', shrink=0.1)
plt.annotate('Max Value', xytext=(0, 0.15), xy=(0, 0.25), arrowprops=props, fontsize=14, ha="center")
plt.title("Derivation of Sigmoid Function", fontsize=14)
plt.savefig('sigmoid_derivation.png')
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
# Set the figure size
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [5, 3]
# Set up with a higher resolution screen
%config InlineBackend.figure_format = 'retina'
plt.style.use('fivethirtyeight')

def leaky_relu(z, alpha=0.01):
    return np.maximum(alpha*z, z)

z = np.linspace(-5, 5, 200)
plt.plot(z, leaky_relu(z, 0.05), "blue", linewidth=3, label="LeakyReLU")
# plt.plot([-5, 5], [0, 0], 'k')
# plt.plot([0, 0], [-0.5, 6.2], 'k')
# plt.grid(True)
props = dict(facecolor='black', shrink=0.1)
plt.annotate('Leak', xytext=(-3.5, 0.5), xy=(-5, -0.2), arrowprops=props, fontsize=14, ha="center")
plt.title("Leaky ReLU activation function", fontsize=14)
# plt.axis([-5, 5, -0.5, 4.2])

plt.savefig("leaky_relu_plot.jpg", dpi=500, bbox_inches='tight')
plt.show()
```
