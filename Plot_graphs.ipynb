```python
# importing package
import matplotlib.pyplot as plt
  
# create data
x = ['A', 'B', 'C', 'D','E','F']
y1 = [23,15,9,12,43,14]
y2 = [463, 123, 151, 111,684,56 ]
  
# plot bars in stack manner
a=plt.bar(x, y1, color='red')
b=plt.bar(x, y2, bottom=y1, color='orange')
for rect in b:
        height = rect.get_height()
#         print(height)
        plt.text(rect.get_x() + rect.get_width()/2.0, height, '    %d ' % float(height), ha='center', va='bottom',rotation=90,size=14,color='black')
plt.show()
```

```python
# importing package
import matplotlib.pyplot as plt
import pandas as pd
plt.figure(figsize=(4,3))
# create data
plt.rcParams["figure.figsize"] = [8, 4]
df = pd.DataFrame([['A', 10, 20, 10, 30], ['B', 20, 25, 15, 25], ['C', 12, 15, 19, 6],
                   ['D', 10, 29, 13, 19]],
                  columns=['Team', 'Round 1', 'Round 2', 'Round 3', 'Round 4'])
# view data
print(df)
plt.figure(figsize=(4,3))
# plot grouped bar chart
df.plot(x='Team',
        kind='bar',
        stacked=False,
        title='Grouped Bar Graph with dataframe')
plt.ylabel('Percentage',size=12)
plt.xlabel("Round",size=12)
plt.savefig('grp.png', dpi=500, bbox_inches='tight')
```
